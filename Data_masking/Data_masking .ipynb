```python
# importing the  libraries
import pandas as pd
import os
```

```python
#read the csv file 
df = pd.read_csv("E:\\Data Masking\\account_detail.csv")
```

```python
#top 5 rows of the data to know thw  datatype  and columns need masking  
df.head()
```

```python
#as both columns need  same type of masking put both in one place 
masking_columns = ['Phone Number','Account Number']
```

```python
#masking Email  column as before @ all character become X
df['Email'] = df['Email'].apply(lambda x: '*' * (x.index('@')) + x[x.index('@'):])
```

```python
#masking phone number and account number as only 4 digit are visible 
def mask_data(data):
    return 'X' * (len(data) - 4) + data[-4:]

for column in masking_columns:
    df[column] = df[column].astype(str).apply(mask_data)
```

```python
#masking applied
df.head()
```

```python
#save the masked output into csv file 
df.to_csv("E:\\Data Masking\\masked.csv",index=False)
```
